```
import requests
import bs4
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```
URL = "https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=Data+Scientist&txtLocation="
#conducting a request of the stated URL above:
page = requests.get(URL)
#specifying a desired format of “page” using the html parser - this allows python to read the various components of the page, rather than treating it as one long string.
soup = BeautifulSoup(page.text, "html.parser")
#printing soup in a more structured tree format that makes for easier reading
#print(soup.prettify())
```

```
def extract_job_title_from_result(soup): 
    jobs = []
    for header in soup.find_all(name="header", attrs={"class":"clearfix"}):
      for a in header.find_all(name="a"):
        jobs.append(a.text.strip())
  
    return(jobs)
extract_job_title_from_result(soup)
```

```
def extract_company_from_result(soup): 
  companies = []
  for h3 in soup.find_all(name="h3", attrs={"class":"joblist-comp-name"}):
    #all_company = div.find_all(name="span", attrs={"class":"joblist-comp-name"})
    #for company in all_company:
      companies.append(h3.text.strip())    
  return(companies)
 
extract_company_from_result(soup)
```

```
def extract_location_from_result(soup): 
  locations = []
  for ul in soup.findAll(name = "ul", attrs={"class": "top-jd-dtl clearfix"}):
    try:
      for span in ul.find_all(name="span"):
        locations.append(span.text.strip())
    except:
      locations.append("Nothing_found")
  return(locations)
extract_location_from_result(soup)
```

```
def extract_summary_from_result(soup): 
  summaries = []
  for ul in soup.findAll(name ="ul", attrs={"class": "list-job-dtl clearfix"}):
      for li in ul.find_all(name="li"):
        summaries.append(li.text.strip())
  return(summaries)
extract_summary_from_result(soup)
```

```
job_title_list = extract_job_title_from_result(soup)
company_list = extract_company_from_result(soup)
location_list = extract_location_from_result(soup)
#salary_list = extract_salary_from_result(soup)
summary_list = extract_summary_from_result(soup)
```

```
import pandas as pd

column_list = ["Job Title", "Company Name", "Location", "Summary"]

data = pd.DataFrame(list(zip(job_title_list, company_list, location_list, summary_list)), columns =column_list)

data
```

```
from tqdm import tqdm

search_page_list = []
for i in range(0,4001,10):
  search_page_list.append(i)

job_title_list = []
company_list = []
location_list = []
salary_list = []
summary_list = []

base_URL = "https://www.timesjobs.com/candidate/job-search.html?from=submit&actualTxtKeywords=Data%20Scientist&searchBy=0&rdoOperator=OR&searchType=personalizedSearch&luceneResultSize=25&postWeek=60&txtKeywords=data%20scientist&pDate=I&sequence=1&startPage="

for page in tqdm(search_page_list):
  URL = base_URL + str(page)
  page = requests.get(URL)
  soup = BeautifulSoup(page.text, "html.parser")

  job_title_list.extend(extract_job_title_from_result(soup))
  company_list.extend(extract_company_from_result(soup))
  location_list.extend(extract_location_from_result(soup))
  summary_list.extend(extract_summary_from_result(soup))
  

column_list = ["Job Title", "Company Name", "Location", "Summary"]

data = pd.DataFrame(list(zip(job_title_list, company_list, location_list, summary_list)), columns =column_list)

data
```
